```python
s_df = pd.read_csv('data/survey_results_public.csv')
```

```python
s_df
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
s_df.sort_values(by='Country', inplace=True)
```

```python
s_df['Country'].head(50)
```

```python
s_df[['Country', 'ConvertedCompYearly']].head(50)
```

```python
s_df.sort_values(by=['Country', 'ConvertedCompYearly'], ascending=[True, False], inplace=True)
```

```python
s_df[['Country', 'ConvertedCompYearly']].head(50)
```

```python
s_df['ConvertedCompYearly'].nlargest(10)
```

```python
s_df['ConvertedCompYearly'].nsmallest(10)
```

```python
s_df[['Country', 'ConvertedCompYearly']].nlargest(10, columns=['ConvertedCompYearly'])
```

```python
s_df[['Country', 'ConvertedCompYearly']].nsmallest(10, columns=['ConvertedCompYearly'])
```

```python
s_df.nlargest(10, 'ConvertedCompYearly')
```

```python
s_df.nsmallest(10, 'ConvertedCompYearly')
```

```python
filt = (s_df['Country'] == "India")
df = s_df[filt]
df[['Country', 'ConvertedCompYearly']].nlargest(10, columns=['ConvertedCompYearly'])
```

```python
df[['Country', 'ConvertedCompYearly']].nsmallest(10, columns=['ConvertedCompYearly'])
```

```python
filt = (s_df['Country'] == "India")
s_df.loc[filt, ['Country', 'ConvertedCompYearly']].nlargest(10, columns=['ConvertedCompYearly'])
```

```python
s_df.loc[filt, ['Country', 'ConvertedCompYearly']].nsmallest(10, columns=['ConvertedCompYearly'])
```
