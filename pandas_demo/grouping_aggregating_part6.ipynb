```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count': 'NumRespondents', 'LanguageHaveWorkedWith': 'NumsKnowsPython'}, inplace=True)
```

```python
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['NumsKnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df
```
